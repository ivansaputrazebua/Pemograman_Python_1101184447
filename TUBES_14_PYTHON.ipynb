```
#pathloss
def cos231(f, loc, ht, hu, d):
  if f in range(150,1501):
    ca=69.55
    cb=26.16
  elif f in range(1501,2001):
    ca=46.3
    cb=33.9
  
  
  if loc == 'urban':
    cm=0
  elif loc == 'suburban':
    cm=-2*((np.log10((f/28)**2))-5.4)
  elif loc == 'open':
    cm=-4.78*(np.log(f))+18.33*(np.log(f))-40.94
  else:
    cm='loc salah'
  

  ahr = (1.1*np.log10(f)-0.7)*hu-(1.56*np.log10(f)-1.8)
  Lp = ca+(cb*np.log10(f))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-cm


  return Lp
```

```
f = 1800
Smin = 300
Smax = 1000
N = 200
d = np.random.randint(Smin,Smax,N)
ht = 80
hu = 1.5
d_sort = d[np.argsort(d)]
d = d_sort
```

```
loc = 'urban'
LpUrban = cos231(f, loc, ht, hu, d)
urban_sort = LpUrban[np.argsort(d)]
loc = urban_sort
```

```
loc = 'suburban'
LpSuburban = cos231(f, loc, ht, hu, d)
suburban_sort = LpSuburban[np.argsort(d)]
loc = suburban_sort
```

```
loc = 'open'
LpOpen = cos231(f, loc, ht, hu, d)
open_sort = LpOpen[np.argsort(d)]
loc = open_sort
```
